```python
import networkx as nx

example = """162,817,812
57,618,57
906,360,560
592,479,940
352,342,300
466,668,158
542,29,236
431,825,988
739,650,466
52,470,668
216,146,977
819,987,18
117,168,530
805,96,715
346,949,466
970,615,88
941,993,340
862,61,35
984,92,344
425,690,689"""


def format(data):
    data =  data.split('\n')
    data = [[int(coordinate)for coordinate in coordinates.split(',')]for coordinates in data]
    return data

with open('eigth.txt', 'r', encoding='utf-8') as file:
    data = file.read()
```

```python
data = format(data)
```

```python
def distance(p,q):
    return sum([(p[i]-q[i])**2 for i in range(3)])**.5
```

```python


distances = [[distance(i,j)for i in data]for j in data]


shortest_distances = []
for i in distances:
    for j in i:
        shortest_distances.append(j)

shortest_distances

shortest_distances_indices = [i[0] for i in sorted(enumerate(shortest_distances), key=lambda x:x[1])]
```

```python
G = nx.Graph()
G.add_nodes_from([str(i)for i in data])

length = len(data)
history = []
c = 0
for index in shortest_distances_indices:     
        index_2 = shortest_distances.index(shortest_distances[index])//length
        index_1 = distances[index_2].index(shortest_distances[index])
        d = distance(data[index_1], data[index_2])
        if d not in history and d!=0:
            history.append(distance(data[index_1], data[index_2]))
            G.add_edge(str(data[index_1]), str(data[index_2]))
            c+=1
            if c==length:
                break

circuits_lenghts = [len(c) for c in sorted(nx.connected_components(G), key=len, reverse=True)]
mult = 1
for i in circuits_lenghts[:3]:
     mult*=i
mult
```
