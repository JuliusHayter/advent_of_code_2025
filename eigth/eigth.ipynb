```python
example = """162,817,812
57,618,57
906,360,560
592,479,940
352,342,300
466,668,158
542,29,236
431,825,988
739,650,466
52,470,668
216,146,977
819,987,18
117,168,530
805,96,715
346,949,466
970,615,88
941,993,340
862,61,35
984,92,344
425,690,689"""


def format(data):
    data =  data.split('\n')
    data = [[int(coordinate)for coordinate in coordinates.split(',')]for coordinates in data]
    return data

```

```python
data = format(example)
```

```python
def distance(p,q):
    return sum([(p[i]-q[i])**2 for i in range(3)])**.5
```

```python
distances = [[distance(i,j)for i in data if i!=j]for j in data]

shortest_distances = [min(i)for i in distances]

shortest_distances_indices = [i[0] for i in sorted(enumerate(shortest_distances), key=lambda x:x[1])]
```
