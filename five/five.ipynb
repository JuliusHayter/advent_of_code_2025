```python
example = """3-5
10-14
16-20
12-18

1
5
8
11
17
32"""


with open('five.txt', 'r', encoding='utf-8') as file:
    data = file.read()


def format(data):
    data = data.split('\n\n')
    data = [i.split('\n')for i in data]
    data[1] = [int(i)for i in data[1]]

    data = data[0]
    data = [i.split('-')for i in data]
    data = [[int(i)for i in j]for j in data]
    data = [[i[0],i[1]+1]for i in data]
    return data

data = format(data)
```

```python
def between(n, range_list):
    return n>=range_list[0] and n<=range_list[1]

def is_common_range(range_one, range_two):
    return (between(min(range_one), range_two)) or (between(min(range_two), range_one))

def merge_range(range_one, range_two):
    return [min(min(range_one), min(range_two)), max(max(range_one), max(range_two))]

def iteration(data):
    for i in data:
        for j in data:
            if i != j:
                if is_common_range(i,j):
                    data.remove(i)
                    data.remove(j)
                    data.append(merge_range(i,j))
                    temp_data = []
                    for i in data:
                        if i not in temp_data:temp_data.append(i)
                    data = temp_data
                    return data
                
def count_all(data):
    while True and len(data)>2:
        length = len(data)
        temp_data =data
        data = iteration(data)
        if data == None:
            data = temp_data
            break
        if len(data) == length:
            break

    count = 0
    for i in data:
        count += i[1]-i[0]
    return count

count_all(data)
```
