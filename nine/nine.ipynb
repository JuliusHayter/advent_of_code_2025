```python
example = """7,1
11,1
11,7
9,7
9,5
2,5
2,3
7,3"""

with open('nine.txt', 'r', encoding='utf-8') as file:
    data = file.read()




def format(data):
    data = data.split('\n')
    data = [[int(pos)for pos in couple.split(',')]for couple in data]
    return data
```

```python
data = format(data)
```

```python
def square_size(a,b):
    side_1 = abs(a[0]-b[0])+1
    side_2 = abs(a[1]-b[1])+1
    return side_1*side_2
```

```python
maxi = 0
for i in data:
    for j in data:
        if i!=j:
            size = (square_size(i,j))
            if size > maxi:
                maxi = size

maxi
```

```python
from shapely.geometry import Polygon


def rectangle(a,b):
    return [a,[b[0],a[1]],b,[a[0], b[1]]]
```

```python
dico = {}
for i in data:
    for j in data:
        if i!=j:
            dico[square_size(i,j)] = [i,j]
```

```python
dico = sorted(dico.items(), reverse=True)
```

```python
for i in dico:
    coords = (i[1])
    if[coords[0][0],coords[1][1]] in data or [coords[1][0], coords[0][1]] in data:
        print(i)
```

```python
data_polygon = Polygon(data)
for i in dico:
    coords = i[1]
    poly = Polygon(rectangle(coords[0],coords[1]))
    if (data_polygon.contains(poly)):
        print(i)
        break
```
