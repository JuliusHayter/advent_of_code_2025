# first part

```python
example = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82"""

example = example.split("\n")

def decrypt(starting_point, code):

    pos = starting_point
    count = 0

    for i in code:
        dir = i[0]
        distance = int(i[1:])
        if dir == "L":
            pos-=distance
        else:
            pos += distance
        pos = pos%100


        if pos == 0:
            count +=1

    return count
```

```python
decrypt(50, example)
```
