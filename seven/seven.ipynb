```python
example = """.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
..............."""

with open('seven.txt', 'r', encoding='utf-8') as file:
    data = file.read()

def format(data):
    return data.split('\n')

data = format(example)
data = [[character for character in row]for row in data]
data
```

```python
def count_splits(data):
    count = 0
    for index, row in enumerate(data):
        tachyons_index = ([index_in_row for index_in_row in range(len(row))if row[index_in_row]in ['S', '|']])
        for tachyon_index in tachyons_index:
            if index != len(row):
                if data[index+1][tachyon_index] == '^':
                    data[index+1][tachyon_index-1] = '|'
                    data[index+1][tachyon_index+1] = '|'
                    count += 1
                else:
                    data[index+1][tachyon_index] = '|'
    return count


count_splits(data)
```

```python
def split(queue,count):
    for index, row in enumerate(queue[0]):
        tachyons_index = ([index_in_row for index_in_row in range(len(row))if row[index_in_row]in ['S', '|']])
        for tachyon_index in tachyons_index:
            if index != len(row):
                if queue[0][index+1][tachyon_index] == '^' and '|' not in queue[0][index+1]:
                    copy_left = [x[:] for x in queue[0]]
                    copy_left[index+1][tachyon_index-1] = '|'
                    count+=1
                    queue.append(copy_left)


                    copy_right = [x[:] for x in queue[0]]
                    copy_right[index+1][tachyon_index+1] = '|'
                    count +=1
                    queue.append(copy_right)

                    queue = queue[1:]

                    queue_without_duplicates = []
                    for i in queue:
                        if i not in queue_without_duplicates:
                            queue_without_duplicates.append(i)

                    return queue, count
                else:
                    queue[0][index+1][tachyon_index] = '|'
            else:
                queue_without_duplicates = []
                for i in queue:
                    if i not in queue_without_duplicates:
                        queue_without_duplicates.append(i)
                return queue[1:], count
```
