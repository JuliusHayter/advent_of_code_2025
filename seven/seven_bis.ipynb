```python
import numpy as np

example = """.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
..............."""

with open('seven.txt', 'r', encoding='utf-8') as file:
    data = file.read()

def format(data):
    return data.split('\n')

data = format(data)
data = [[character for character in row]for row in data]
data
```

```python
def count_splits(data):
    count = 0
    for index, row in enumerate(data):
        tachyons_index = ([index_in_row for index_in_row in range(len(row))if row[index_in_row]in ['S', '|']])
        for tachyon_index in tachyons_index:
            if index != len(row):
                if data[index+1][tachyon_index] == '^':
                    data[index+1][tachyon_index-1] = '|'
                    data[index+1][tachyon_index+1] = '|'
                    count += 1
                else:
                    data[index+1][tachyon_index] = '|'
    return count


count_splits(data)
```
