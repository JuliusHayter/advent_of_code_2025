```python
example = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   + """
example = example.split('\n')

with open('six.txt', 'r', encoding='utf-8') as file:
    data = file.read()
data = data.split('\n')

def load(data): 
    r = [""]*len(data[:-1])

    for i in range(len(data[0])):
        lil_col = [data[j][i]for j in range(len(data[:-1]))]
        if lil_col != [' ']*len(data[:-1]):
            for index, character in enumerate(lil_col):
                if character == ' ':
                    lil_col[index] = "X"
        for index, character in enumerate(lil_col):
            r[index]+= character

    r = [i.split()for i in r]
    r.append(data[-1].split())
    return r


data = load(data)
data
```

```python
def transfo(data):    
    data = [i[::-1]for i in data]
    max_length = max([len(i)for i in data])
    transformed_data = [""]*max_length
    for index,number in enumerate(data):
        for i in range(max_length):
            if number[i] != 'X':
                transformed_data[i] += number[i]
    return (transformed_data)

transfo(['123', 'X45', 'XX6'])
```

```python
def calculate(data):
    grand_total = 0
    for index in range(len(data[0])):
        operator = data[-1][index]
        column = [i[index]for i in data[:-1]]
        print("^^^",column)
        column = transfo(column)
        print(column, operator)
        grand_total += eval((operator.join(column)))
    return grand_total

calculate(data)
```
