```python
def is_symetric(n):
    n = str(n)
    l = len(n)
    return n[:l//2] == n[l//2:] 

def find_invalid_ids(data):
    count = 0
    for id_range in data:
        all_range = [ids for ids in range(id_range[0], id_range[1]+1)]
        symetrics = [ids for ids in all_range if is_symetric(ids)]
        for i in symetrics:
            count += i

    return count
```

```python
find_invalid_ids(example)
```

```python
def find_invalid_ids(data):
    count = 0
    for id_range in data:
        all_range = [ids for ids in range(id_range[0], id_range[1]+1)]
        candidates = [str(i) for i in all_range if len(set(str(i)))<=5]

        for candidate in candidates:
            j=1
            length = len(candidate)
            while j<length:
                if length%j==0:
                    if candidate[:j]*(length//j) == candidate:
                        count+= int(candidate)
                        break
                j+=1

    return count
```

```python
find_invalid_ids(data)
```
